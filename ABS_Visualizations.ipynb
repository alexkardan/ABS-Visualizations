```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
data = pd.read_csv('ABS_Sales_special items.csv')
```

```python
data.head(7)
```

```python
plt.figure(figsize = (10,5))
data["Category"].value_counts().head(10)[::-1].plot(kind = "barh", color = 'blue', alpha = 0.8)
plt.title("Alcohol types sold by ABS")
plt.xlabel("Number of Items in each Category")
plt.ylabel("Alcohol Types")
plt.show()
```

```python
data.groupby("Category")["Total Inventory"].sum().sort_values(ascending=False).head(10)[::-1].plot(kind = "barh", color = 'black', alpha = 0.8)
plt.ylabel("Alcohol Types")
plt.xlabel("Inventory amounts")
plt.title("Alcohol Inventory Levels")
plt.grid(True, color = 'grey', alpha = 0.5)
plt.xticks(rotation = 35)
plt.show()
```

```python
plt.figure(figsize = (10,6))
data["Discount Amount"] = data["Price"] - data["Sale Price"]
data["Discount %"] = (data["Discount Amount"]/ data["Price"]) * 100
plt.scatter(data["Total Inventory"], data["Discount %"], color = "black", alpha = 0.8)
plt.xlabel("Total Inventory")
plt.ylabel("Discount Percentage (%)")
plt.title("Are Lower Inventory Levels More Likely to Have Bigger Discounts?")
plt.grid(True, linestyle = "-.", color = 'blue', alpha = 0.7)
plt.show()
```

```python
data.groupby("Category")["Total Inventory"].sum().sort_values(ascending=False).head(20).plot(
    kind="bar", color="blue", alpha=0.8, figsize=(10,5)
)

plt.ylabel("Total Inventory")
plt.xlabel("Category")
plt.title("Top 20 Categories by Total Inventory")
plt.xticks(rotation=45)  # Rotate x-axis labels for readability
plt.show()
```
